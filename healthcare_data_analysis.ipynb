# How much is Health Care costing American Businesses?
## An analysis of premium cost in the United States and Comparison of Healthcare Cost of Countries with Single-payer Healthcare Systems. 


```
import pandas as pd
import plotly.offline as py
import plotly.graph_objects as go

plus_one = pd.read_csv('Prem_Cost_Employee_plus_One.csv', skiprows=2, skipfooter=14, engine='python')
family = pd.read_csv('Prem_Cost_Family.csv', skiprows=2, skipfooter=12, engine='python')
single = pd.read_csv('Prem_Cost_Single_Employee.csv', skiprows=2, skipfooter=12, engine='python')
```

```
import plotly.graph_objects as go

converter = lambda amount, currency: "%s%s%s" %(
    "-" if amount < 0 else "", 
    currency,('{:%d,.2f}'%
              (len(str(amount))+3)).format(abs(amount)).lstrip()) # To format values as currency

headerColor = 'darkblue'
rowEvenColor = 'lightgrey'
rowOddColor = 'white'

fig = go.Figure(data=[go.Table(
    header=dict(values=['Plan Type', 'Employee Contribution', 'Employer Contribution', 'Total Annual Premium Cost'],
                line_color='darkslategray',
                fill_color= headerColor,
                align=['left', 'center'],
                font=dict(color='white', size=14)),
    cells=dict(values=[['Single Employee', 'Employee Plus One', 'Family'], # 1st column
                       [converter(single.iloc[0,16], '$'), 
                        converter(plus_one.iloc[0,16], '$'), 
                        converter(family.iloc[0,16], '$')], # 2nd column
                       [converter(single.iloc[0,17], '$'), 
                        converter(plus_one.iloc[0,17], '$'), 
                        converter(family.iloc[0,17], '$')], # 3rd column
                       [converter(single.iloc[0,18], '$'), 
                        converter(plus_one.iloc[0,18], '$'), 
                        converter(family.iloc[0,18], '$')]], # 4th column
               line_color='darkslategray',
               fill_color= [[rowOddColor,rowEvenColor,rowOddColor, rowEvenColor,rowOddColor]*4],
               align=['left', 'center']))
])

fig.update_layout(
    title_text = 
    '2018 Average Annual Premium Per Enrolled Employee')

fig.show()
```

```
states = ["AL", "AK", "AZ", "AR", "CA", "CO", "CT", "DE", "DC", "FL", "GA", 
          "HI", "ID", "IL", "IN", "IA", "KS", "KY", "LA", "ME", "MD", 
          "MA", "MI", "MN", "MS", "MO", "MT", "NE", "NV", "NH", "NJ", 
          "NM", "NY", "NC", "ND", "OH", "OK", "OR", "PA", "RI", "SC", 
          "SD", "TN", "TX", "UT", "VT", "VA", "WA", "WV", "WI", "WY"]
```

```
family_1 = family.drop(labels=[0])
family_1['code'] = states

for col in family_1.columns:
    family_1[col] = family_1[col].astype(str)

family_1['text'] = family_1['Location'] + '<br>' + \
  'Employer Contribution:$' + family_1['2018__Employer Contribution'] + '<br>'+ \
  'Employee Contribution:$' + family_1['2018__Employee Contribution']

family_1.head()
```

```
fig1 = go.Figure(data=go.Choropleth(
    locations=family_1['code'],
    z = family_1['2018__Total Annual Premium'].astype(float),
    locationmode = 'USA-states',
    colorscale = 'Blues',
    autocolorscale=False,
    text=family_1['text'],
    colorbar_tickprefix = '$',
    colorbar_title="Premiums USD"   
))

fig1.update_layout(
    title_text = '2018 Average Annual Family Premium per Enrolled Employee',
    geo_scope='usa',
)


fig1.show()
```
